###Análise da expectativa de vida para 2025$$$

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FILE = '/content/overall_life_expectancy_2025.csv'
df = pd.read_csv(FILE, sep=',')
df.head(10)
```

***Excluindo colunas***

```python
df.drop(['id','expectancy'], axis=1, inplace=True)
```

***Trocando os nomes das colunas***

```python
df.columns = ['País', 'Expectativa', 'Continente']
```

```python
df.head()
```

***Expectativa de vida média nos continentes***

```python
mean_continent = df.groupby('Continente')['Expectativa'].mean().sort_values(ascending=False)
```

```python
print(mean_continent)
```

```python
sns.set_style('darkgrid')
sns.barplot(x=mean_continent.values, y=mean_continent.index, color='blue')
plt.title('Média da Expectativa de Vida nos Continentes')
```

***País com maior expectativa de vida para cada continente***

```python
country_max = df.loc[df.groupby('Continente')['Expectativa'].idxmax()]
```

```python
print(country_max)
```
